```python
from services.aqi_agent.src.aqi_agent.domain.table_pruner import TablePrunerService, TablePrunerInput
from opensearch import OpenSearchService
from lite_llm import LiteLLMService
from services.aqi_agent.src.aqi_agent.shared.utils import get_settings

settings = get_settings()

litellm_service = LiteLLMService(settings = settings.litellm)
open_search_service = OpenSearchService(settings = settings.opensearch)
```

```python
import nest_asyncio
import nest_asyncio

nest_asyncio.apply()


input = TablePrunerInput(
    question = "Cầu Giấy hôm nay thế nào"
)
table_pruner_service = TablePrunerService(
    opensearch_service = open_search_service,
    litellm_service = litellm_service,
    table_pruner_settings = settings.table_pruner
)

result = await table_pruner_service.process(inputs = input)
```

```python
result_json = result.model_dump()

import json

with open("result.json", "w", encoding = "utf-8") as f:
    json.dump(result_json, f)
```
