```python
import sqlglot

sql_query = """
SELECT 
    u.id,
    u.name,
    u.age,
    d.department_name,
    COUNT(o.id) AS total_orders,
    AVG(o.amount) AS avg_order_amount
FROM users u
JOIN departments d 
    ON u.department_id = d.id
LEFT JOIN orders o 
    ON u.id = o.user_id
WHERE 
    u.age > 30
    AND d.location = 'Hanoi'
    AND u.id IN (
        SELECT user_id
        FROM orders
        WHERE amount > 100
    )
GROUP BY 
    u.id, u.name, u.age, d.department_name
HAVING 
    COUNT(o.id) > 5
ORDER BY 
    avg_order_amount DESC
LIMIT 10
"""
parsed_statements = sqlglot.parse(sql_query)
```
